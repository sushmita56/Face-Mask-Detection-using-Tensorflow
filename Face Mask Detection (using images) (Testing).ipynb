```python
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
import numpy as np
import cv2
import os
```

```python
model = load_model('D:\Deep Learning\FaceMaskDetector\maskdetector.model')
```

```python
model.summary()
```

```python
prototxtPath = os.path.sep.join([r'D:\Deep Learning\FaceMaskDetector\face detector', 'deploy.prototxt'])
weightsPath = os.path.sep.join([r'D:\Deep Learning\FaceMaskDetector\face detector', 'res10_300x300_ssd_iter_140000.caffemodel'])
```

```python
weightsPath
```

```python
#Reads the network model (stored in two variables: prototxtPath, weightsPath) in Caffe framework's format.
faceNet = cv2.dnn.readNet(prototxtPath, weightsPath) # facedetector
```

```python
faceNet
```

```python
img = cv2.imread('D:\Deep Learning\FaceMaskDetector\images\image-4.jpeg')
```

```python
img
```

```python
(h,w) = img.shape[:2]
```

```python
(h,w)
```

```python
#get our blob which is our input image
blob = cv2.dnn.blobFromImage(img, 1.0, (300,300), (104.0, 177.0, 123.0))
```

```python
blob
```

```python
blob.shape
```

```python
#input the blob into the model and get back the detections 
faceNet.setInput(blob)
detections = faceNet.forward()
```

```python
detections[0]
```

```python
print(detections.shape)
print(detections.shape[2])
```

```python
detections[0,0,:,2]
```

```python
for i in range(0, detections.shape[2]):
    confidence = detections[0, 0, i, 2]
    
# Extract confidence and compare it to the confidence threshold(50%) so that we can filter out detections that are weak.
# If the algorithm is more than 50% confident that the detection is a face, Iterate over all of the faces detected and extract their start and end points.

    if (confidence > 0.5):
        box = detections[0, 0, i, 3:7]*np.array([w,h,w,h])
        startX, startY, endX, endY = box.astype('int')
        
        # ensure the bounding boxes fall within the dimensions of the frame
        startX, startY = (max(0, startX), max(0,startY))
        endX, endY = (min(w-1, endX), min(h-1, endY))
        
        # extract the face ROI, convert it from BGR to RGB channel, resize it to 224,224 and preprocess it
        face = img[startY:endY, startX:endX]
        face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
        face = cv2.resize(face, (224,224))
        face = img_to_array(face)
        face = preprocess_input(face)
        face = np.expand_dims(face, axis=0)
        
        withMask, withoutMask = model.predict(face)[0]
        
        # determine the class label and color(the class label and the bounding box will be shown in the specified color)
        label = "Mask" if withMask > withoutMask else "No Mask"
        color = (0,255,0) if label=='Mask' else (0,0,255)
        
        label="{}: {:.2f}%".format(label,max(withMask,withoutMask)*100)
 
   
# If the algorithm is more than 50% confident that the detection is a face, show a rectangle around it .
# display the label(mask or withoutmask) and rectagular bounding boxes around the face
        cv2.putText(img, label, (startX, startY-10), cv2.FONT_HERSHEY_SIMPLEX, 0.45, color, 2)
        cv2.rectangle(img, (startX, startY), (endX, endY), color, 2)
        

cv2.imshow("Output", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
